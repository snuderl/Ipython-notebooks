# B-kromatično število in krepki produkt

B-barvanje je pravilno barvanje grafa G za katerega velja, da ima vsaka barva vsaj eno vozlišče, ki ima soseda vseh ostalih barv. B-kromatično število je največje celo število $ \varphi(x)$ za katero ima graf G b-barvanje. Trivialna spodnja meja $ \varphi(x) $ je kromatično število $ \chi (x) $, zgornja meja pa $ \Delta (x) $ + 1.

**Izrek 1:** Naj bo $ v_i \in V(G)$ in  $ d(v_1) \geq d(v_2) \geq ... \geq d(v_n)$. Število m(G) = $ max\{ i | d(v_i) \geq i - 1\} $.

Izrek je zelo intuitiven in pravi da za k-b-barvanje potrebujemo vsaj k vozlišč z k - 1 sosedi.

## Krepki produkt

Krepki produkt grafov je binarna operacija $ \boxtimes $ nad dvema grafoma. Vozlišla $ G \boxtimes H $ so vsi pari $ V(G) \times V(H) $. Vozlišči (u, v) in (u', v') v krepkem produktu pa sta sosednji:
    * u = u' in v je sosednji z v' ali,
    * v = v' in u je sosednji z v' ali,
    * u je sosednji z u' in v je sosednji z v'.

## Spodnja meja

**Izrek 2:** Naj bosta G in H poljubna grafa. Potej je $ \varphi( G \boxtimes H) \geq \varphi(G) * \varphi(H) $.

**Dokaz.**
Naj bo $ c_1 \varphi(G)$-b-barvanje grafa G in $ c_2 \varphi(H)$-b-barvanje grafa H. Definirajmo barvanje $c_3$ na grafu $ G \boxtimes H $ kot $$c_3((u, v)) = (c_1(u), c_2(v)) $$.

Očitno je da je barvanje $ c_3 $ pravilno barvanje. Pokazati je treba da je tudi b-barvanje. Naj bo u poljubno vozlišče barve i, ki realizira vse barve v G in v poljubno vozlišče barve j, ki realizira vse barve v H. Barva vozlišča (u, v) v produktu je potem (i, j). Vozlišči u in v sta sosednja vsem barvam razen svojim, zato mora biti tudi produkt po definiciji soseden vsem tem barvam. Ker je u barve i, je vozlišče prav tako sosednje barvam (i, $x_k$) in podobno za v.

## Zgornja meja

B-kromatično število produkta lahko navzgor omejimo s pomočjo trivialne zgornje meje. $$ \varphi( G \boxtimes H) \leq \Delta(G \boxtimes H) + 1 = \Delta(G)\Delta(H) + \Delta(G) + \Delta(H) + 1 $$.

Enakost dobimo če sta b-kromatični števili grafov G in H že zgornji meji. Naj bo $ \varphi( G ) = \Delta(G) + 1 $ in $ \varphi(H) = \Delta(H) + 1 $ potem sledi da je $$ \varphi( G \boxtimes H) = \varphi(G)\varphi(H) $$.

Nasprotna trditev ne drži vedno kar je razvidno npr. iz produkta $ P_5 $ in $ P_3 $.

**Izrek 3:** Naj bo H poljuben graf. Potem je $ \varphi(P_n \boxtimes H) = 3(\Delta(H) + 1) = \varphi(C_n \boxtimes H) $ za vsak 
    $ n \geq 3(\Delta(H) + 1) + 2$.
    
**Dokaz.** Naj bojo $C_i $ i=1,2,3 3 skupine barv moči $\Delta(H) + 1$, ki imajo prazen presek. Naj bo u vozlišče maksimalne stopnje v grafu H in $ v_i $ vozlišče v $ P_n $. Poglejmo si vlakna $ H^{v_i} $. Vozlišče z drugo koordinatu u v tem vlaknu ima $ \Delta(H) $ sosedov v vlaknu. Vlakno lahko tako prebarvamo z $ C_1 $. Vlakni $ H^{v_{i-1}} $ in $ H^{v_{i+1}} $ pa lahko prebarvamo z barvami C2 in C3. Pri barvanju posameznega vlakna vozlišče $ v_i $, prebarvamo z i/3-to barvo v ustreznem naboru barv C1, C2 ali C3. Barve v posameznih vlaknih si tako sledijo v zaporedju C1, C2, C3, C1, C2... Ker je u maksimalno vozlišče, znotraj vlakna \Delta{H} sosedov razlinčnih barv + 1 za svojo barvo. Prav tako ima \Delta{H} + 1 sosedov na vlaknu višje in podobno nižje. Na pozicijah ($v_i$, u) so tako vozliščna sosedna z vsemi barvami. Ker je $ n \geq 3(\Delta(H) + 1) + 2$ je pozicij ($v_i$, u) dovolj, da se na njih vsaj enkrat pojavi vsaka barva.

Enakost sledi iz zgornje meje v prejšnem odseku, kjer je $ \Delta{G} = 2 $.

```python
import matplotlib.pyplot as plt
import networkx as nx

P3 = nx.path_graph(3)
G = nx.algorithms.operators.product.strong_product(P3, P3)
pos=nx.spring_layout(G)
nodes = [[(0,0), (1, 2), (2, 0)], [(1,1)], [(0,1), (2, 1)], [(0, 2)], [(1, 0), (2, 2)]]
colors = ["r", "b", "g", "y", "silver"]

for node, color in zip(nodes, colors):    
    nx.draw_networkx_nodes(G,pos,
                           nodelist=node,
                           node_color=color,
                           node_size=500,
                       alpha=0.8)

title(r"$ P_3 \boxtimes  K_{1,2} $", fontsize=18)
axis("off")
nx.draw_networkx_edges(G,pos,width=1.0,alpha=0.5)
```

```python
import matplotlib.pyplot as plt
import networkx as nx

P4 = nx.path_graph(4)
star = nx.star_graph(1)
G = nx.algorithms.operators.product.strong_product(P4, star)
pos=nx.spring_layout(G)
nodes = [[(0,0), (2, 0)], [(1,1), (3, 1)], [(0,1), (2, 1)], [(1, 0), (3, 0)]]
colors = ["r", "b", "g", "y"]

for node, color in zip(nodes, colors):    
    nx.draw_networkx_nodes(G,pos,
                           nodelist=node,
                           node_color=color,
                           node_size=500,
                       alpha=0.8)

title(r"$ P_4 \boxtimes  K_{1,1} $", fontsize=18)
axis("off")
nx.draw_networkx_edges(G,pos,width=1.0,alpha=0.5)
```
